```python
# a principo vamos exibir as MEEDIAS MEDIANAS E DEESVIOS-PADRAO das colunas numericas
#separarei em um novo dataframee (dados_numeericos)e tratarei as respostas q mee pareceem q nao seerao eexibidas numeericameente

dados_numericos = arquivo.loc[:, "restricao_vocal_1":"recuperacao_vocal_3"]

dados_numericos.head()
```

```python
#criando um dicionario para substituir os valores q sao strings para numeericos
def valores_inteiros (valor_string):
    if valor_string == "0 = nunca":
        return 0
    elif valor_string == "1 = quase nunca":
        return 1
    elif valor_string == '2 = às vezes':
        return 2
    elif valor_string == "3 = quase sempre":
        return 3
    elif valor_string == "4 = sempre":
        return 4
    elif isinstance(valor_string, int):
        return valor_string
    else:
        return 100000

for coluna in dados_numericos.columns:
    for linha in dados_numericos.index:
        dados_numericos.at[linha,coluna]= valores_inteiros(dados_numericos.at[linha,coluna])

dados_numericos.head()
```
