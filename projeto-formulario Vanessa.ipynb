```python
# vou simplificar o nome desssas colunas q eestao me deixando confusas 
# para isso vou usar palavras chavees ou expressoees nas peerguntas

novos_nomes = ["data", "termo", "email", "inicias", "nascimento", "genero", "altura","peso", "area_atuacao",
               "freq_voos","media_duracao_voos", "tempo_aviacao", "sintomas", "orientacoes", "importancia_voz",
               
               "restricao_vocal_1", "restricao_vocal_2", "restricao_vocal_3","restricao_vocal_4","restricao_vocal_5",
               "restricao_vocal_6","restricao_vocal_7","restricao_vocal_8", "restricao_vocal_9", "restricao_vocal_10", "restricao_vocal_11",
               
               "desconforto_fisico_1","desconforto_fisico_2","desconforto_fisico_3","desconforto_fisico_4","desconforto_fisico_5",
               
               "recuperacao_vocal_1","recuperacao_vocal_2","recuperacao_vocal_3"] 

arquivo.columns=novos_nomes
```

```python
# a principo vamos exibir as MEEDIAS MEDIANAS E DEESVIOS-PADRAO das colunas numericas
#separarei em um novo dataframee (dados_numeericos)e tratarei as respostas q mee pareceem q nao seerao eexibidas numeericameente

dados_numericos = arquivo.loc[:, "restricao_vocal_1":"recuperacao_vocal_3"]

dados_numericos.head()
```

```python
#dados_numericos.median()

#dados_numericos.mean()

#dados_numericos.std()
```

```python
#criando um dicionario para substituir os valores q sao strings para numeericos
def valores_inteiros (valor_string):
    if valor_string == "0 = nunca":
        return 0
    elif valor_string == "1 = quase nunca":
        return 1
    elif valor_string == '2 = às vezes':
        return 2
    elif valor_string == "3 = quase sempre":
        return 3
    elif valor_string == "4 = sempre":
        return 4
    elif isinstance(valor_string, int):
        return valor_string
    else:
        return 100000

for coluna in dados_numericos.columns:
    for linha in dados_numericos.index:
        dados_numericos.at[linha,coluna]= valores_inteiros(dados_numericos.at[linha,coluna])

dados_numericos.head()
```

# -------------------- Analisando os dados --------------------

```python
# media desvio e mediana para os campos unificados de restricao_vocal, desconforto_fisico e recuperacao_vocal

restricao = dados_numericos.loc[:,"restricao_vocal_1":"restricao_vocal_11"].sum(axis=1)
desconforto = dados_numericos.loc[:,"desconforto_fisico_1":"desconforto_fisico_5"].sum(axis=1)
recuperacao = dados_numericos.loc[:,"recuperacao_vocal_1":"recuperacao_vocal_3"].sum(axis=1)
total = dados_numericos.loc[:,"restricao_vocal_1":"recuperacao_vocal_3"].sum(axis=1)
#restricao_media  = restricao.stack().mean()   meeelhor do jeito q fiz embiaxo com numpy
restricao_media, restricao_mediana, restricao_desvio = np.mean(restricao.values), np.median(restricao.values), np.std(restricao.values)
desconforto_media, desconforto_mediana, desconforto_desvio = np.mean(desconforto.values), np.median(desconforto.values), np.std(desconforto.values)
recuperacao_media, recuperacao_mediana, recuperacao_desvio = np.mean(recuperacao.values), np.median(recuperacao.values), np.std(recuperacao.values)
total_media, total_eediana, total_desvio = np.mean(total.values), np.median(total.values), np.std(total.values)
restricao_desvio,desconforto_desvio, recuperacao_desvio, total_desvio= round(restricao_desvio,2), round(desconforto_desvio,2), round(recuperacao_desvio,2), round(total_desvio,2)

print ("RESTRICAO\nMedia=", restricao_media, "\nMediana=", restricao_mediana, "\nDesvio padrao=", restricao_desvio)
print ("\nDESCONFORTO\nMedia=", desconforto_media, "\nMediana=", desconforto_mediana, "\nDesvio padrao=", desconforto_desvio)
print ("\nRECUPERACAO\nMedia=", recuperacao_media, "\nMediana=", recuperacao_mediana, "\nDesvio padrao=", recuperacao_desvio)
print  ("\nTOTAL\nMedia+", total_media, "\nMediana=", total_eediana, "\nDesvio padrao=", total_desvio)
```

```python
# calculando os quartis de .25 e .75 ( 25% e 75% respeectivameente)
restricao_quantis = restricao.quantile([.25,.75])
desconforto_quantis = desconforto.quantile([.25,.75])
recuperacao_quantis = recuperacao.quantile([.25,.75])
total_quqntis= total.quantile([.25,.75])

#restricao_quantis, desconforto_quantis, recuperacao_quantis, total_quqntis
```

```python
data = { "Restrição Vocal":[restricao_media, restricao_mediana,restricao_desvio] ,
         "Desconforto Físico":[desconforto_media, desconforto_mediana, desconforto_desvio], 
         "Recuperação vocal":[recuperacao_media, recuperacao_mediana, recuperacao_desvio],
         "Total":[total_media, total_eediana, total_desvio]}
resultados = pd.DataFrame(data,index=["Média","Mediana","Desvio Padrao"]).transpose()
resultados.insert(2,"25%", [restricao_quantis.iloc[0],desconforto_quantis.iloc[0],recuperacao_quantis.iloc[0],total_quqntis.iloc[0]])
resultados.insert(3, "75%", [restricao_quantis.iloc[1],desconforto_quantis.iloc[1],recuperacao_quantis.iloc[1],total_quqntis.iloc[1]])

resultados
```
